```python
#Removing or Imputing missing and Null values
df = df[~df['customer_name'].isnull() & (df['customer_name'].str.strip() != '')]
df = df[~df['customer_segment'].isnull() & (df['customer_segment'].str.strip() != '')]
df['order_priority'].fillna('Not Specified', inplace=True)
df= df[df['unit_price']!='not found']
df = df[df['product_base_margin'].notnull()]
```

```python
# Checking outliers and Percentage of Profit margin by product category
sns.boxplot(data=df, x='product_category', y='profit_margin')
```

While Technology drives the highest revenue with a strong margin (41.5%), Furniture stands out with the highest profit margin (45.3%), making it a high-value category with growth potential. In contrast, Office Supplies contributes the least to both sales and margin (31.7%), so we have to dive deeper. 

```python

#Calculate IQR (interquartile range) for each category
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
 #Apply outlier removal grouped by product category
df = df.groupby('product_category', group_keys=False).apply(
    lambda group: remove_outliers_iqr(group, 'profit_margin')
)

# Plot the boxplot without outliers
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='product_category', y='profit_margin')
plt.title('Profit Margin by Product Category (Outliers Removed)')
plt.show()
```

```python
# Profit Margin Based on Region
grouped_df = df.groupby('region').agg({
    'sales': 'sum',
    'profit_margin': 'mean'
}).reset_index()

print(grouped_df)
```

The consistent profit margins across regions (~36.7%) suggest a well-controlled pricing and cost structure nationwide. This reduces the likelihood of region-specific margin erosion and shifts the analytical focus toward product categories, discount strategies, and customer segments.

```python
# Profit Margin Based on Customer Segment
grouped_df = df.groupby('customer_segment').agg({
    'sales': 'sum',
    'profit_margin': 'mean'
}).reset_index()

print(grouped_df)
```

Across all customer segments, profit margins remain tightly aligned around the business average (~36.5%), suggesting a consistent pricing strategy and product mix.

```python
# Create shipping cost bins (adjust bin size as needed)
bin_edges = range(0, int(df['shipping_cost'].max()) + 100, 100)
df['shipping_bin'] = pd.cut(df['shipping_cost'], bins=bin_edges)
```

```python
# By product category
df_cat = df.groupby(['shipping_bin', 'product_category'])['profit_margin'].mean().reset_index()

# By region
df_region = df.groupby(['shipping_bin', 'region'])['profit_margin'].mean().reset_index()

# By customer segment
df_segment = df.groupby(['shipping_bin', 'customer_segment'])['profit_margin'].mean().reset_index()
```

```python
# Function to get midpoint of shipping bin
def get_bin_midpoint(b):
    return b.left + (b.right - b.left)/2

# Apply to all three DataFrames
df_cat['shipping_mid'] = df_cat['shipping_bin'].apply(get_bin_midpoint)
df_region['shipping_mid'] = df_region['shipping_bin'].apply(get_bin_midpoint)
df_segment['shipping_mid'] = df_segment['shipping_bin'].apply(get_bin_midpoint)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Avg Profit Margin vs Binned Shipping Cost", fontsize=16)

# By Product Category
sns.lineplot(data=df_cat, x='shipping_mid', y='profit_margin', hue='product_category', ax=axes[0])
axes[0].set_title("By Product Category")
axes[0].set_xlabel("Shipping Cost (bin midpoint)")
axes[0].set_ylabel("Profit Margin")
axes[0].set_ylim(0)

# By Region
sns.lineplot(data=df_region, x='shipping_mid', y='profit_margin', hue='region', ax=axes[1])
axes[1].set_title("By Region")
axes[1].set_xlabel("Shipping Cost (bin midpoint)")
axes[1].set_ylabel("")
axes[1].set_ylim(0)

# By Customer Segment
sns.lineplot(data=df_segment, x='shipping_mid', y='profit_margin', hue='customer_segment', ax=axes[2])
axes[2].set_title("By Customer Segment")
axes[2].set_xlabel("Shipping Cost (bin midpoint)")
axes[2].set_ylabel("")
axes[2].set_ylim(0)

plt.tight_layout(rect=[0, 0.05, 1, 0.95])
plt.show()
```

Insight: Focus on reducing shipping costs or bundling strategies for Technology and Office Supplies.
